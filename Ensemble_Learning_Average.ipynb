Importing Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.linear_model import LinearRegression
```

Load the train dataset from salary_data.csv file

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/salary_data.csv")
```

Separate target and train data

```python
target = df.iloc[:, 1].values
train = df.iloc[:, 0].values
```

Split the train data into training and validation datasets

```python
X_train, X_test, y_train, y_test = train_test_split(train, target, test_size=0.20)
```

```python
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
```

Initialize the model objects with default parameters

```python
model_1 = LinearRegression()
model_2 = xgb.XGBRegressor()
model_3 = RandomForestRegressor()
```

Train all the models on the training dataset

```python
model_1.fit(X_train, y_train)
model_2.fit(X_train, y_train)
model_3.fit(X_train, y_train)
```

Predict the output on the validation dataset

```python
pred_1 = model_1.predict(X_test)
pred_2 = model_2.predict(X_test)
pred_3 = model_3.predict(X_test)
```

Calculate the final prediction by averaging the predictions of all 3 models

```python
pred_final = (pred_1 + pred_2 + pred_3) / 3.0
print("Average Result:", pred_final)
```

Print the mean squared error between the real value and predicted value

```python
print("MSE in Average method:", mean_squared_error(y_test, pred_final))
```
