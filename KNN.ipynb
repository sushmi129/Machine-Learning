Importing Libraries

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import random
```

Load Iris dataset

```python
iris = load_iris()
```

Get list of target names

```python
target = iris.target_names
```

Display sample data from Iris dataset

```python
rn = random.randint(0, 149)
print("Sample Data from Iris Dataset")
print(iris.data[rn], "===>", target[iris.target[rn]])
```

Create feature and target arrays

```python
X = iris.data
y = iris.target
```

Split into training and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Print dataset lengths
print("The Training dataset length:", len(X_train))
print("The Testing dataset length:", len(X_test))

try:
    nn = int(input("Enter number of neighbors: "))
    knn = KNeighborsClassifier(nn)
    knn.fit(X_train, y_train)

    # Display the score
    print("The Score is:", knn.score(X_test, y_test))

    # Get test data from the user
    test_data = input("Enter Test Data: ").split(",")
    test_data = list(map(float, test_data))

    # Predict output
    predicted_output = knn.predict([test_data])

    print("\nPredicted output is:", target[predicted_output])

except ValueError:
    print("Please supply valid input...")
```
