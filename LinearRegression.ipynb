Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

Loading dataset

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/salary_data.csv')
```

Visualizing the changes in variables

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('Experience')
plt.ylabel('Salaries')
plt.show()
```

Dividing the data into Independent and Dependent variables

```python
X = df[['YearsExperience']]  # Independent
y = df[['Salary']]  # Dependent
```

Spliting data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

Fitting the Simple Linear Regression model to the training dataset

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

Prediction of Test and Training set result

```python
y_pred = regressor.predict(X_test)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

Evaluate the algorithm

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

Visualizing the results

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title("Salary vs Experience (Training set)")
plt.xlabel("Years of experience")
plt.ylabel("Salaries")
plt.show()
```

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title("Salary vs Experience (Testing set)")
plt.xlabel("Years of experience")
plt.ylabel("Salaries")
plt.show()
```
